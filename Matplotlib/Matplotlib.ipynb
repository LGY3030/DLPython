```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(0,6,0.1)
b=np.sin(a)
plt.plot(a,b)
plt.show()
```

```python
a=np.arange(0,6,0.1)
b=np.sin(a)
c=np.cos(a)
plt.plot(a,b,label='sin')
plt.plot(a,c,linestyle="--",label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin&cos')
plt.legend()
plt.show()
```
